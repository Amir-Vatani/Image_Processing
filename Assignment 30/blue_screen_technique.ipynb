```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
front = cv2.imread('img/SuperMan.jpg')
front_rgb = cv2.cvtColor(front, cv2.COLOR_BGR2RGB)
plt.imshow(front_rgb)
```

```python
back = cv2.imread('img/city.jpg')
back = cv2.cvtColor(back, cv2.COLOR_BGR2RGB)
plt.imshow(back)
```

```python
col, row, _ = front.shape
back = cv2.resize(back, (row, col))
```

```python
front_hsv = cv2.cvtColor(front, cv2.COLOR_BGR2HSV)
H, S, V = cv2.split(front_hsv)
```

```python
for i in range(col):
    for j in range(row):
        if 36<H[i, j]<80 and 30<S[i, j] :
            front_rgb[i, j] = back[i,j]
```
